```python
import pickle
import os
import sys

sys.path.append('clustering')

from utils import get_data_dir
```

```python
with open(os.path.join(get_data_dir(), 'cluster_df_list.pkl'), 'rb') as f:
    cluster_df_list = pickle.load(f)
```

```python
cluster_df_list[-1]
```

```python
cluster_df_list[-1]['VID'] = cluster_df_list[-1]['VID'].astype(int)
cluster_df_list[-1]['cluster_label'] = cluster_df_list[-1]['cluster_label'].astype(int)
cluster_df_list[-1].rename(columns={'cluster_label':'movMF_label'}, inplace=True)
cluster_df_list[-1].rename(columns={'closest_center_distance':'movMF_distance'}, inplace=True)
cluster_df_list[-1].drop(columns=['sim_score'], inplace=True)
cluster_df_list[-1].set_index('VID', inplace=True)
cluster_df_list[-1]
```

```python
with open(os.path.join(get_data_dir(), "2010s", 'wv'), 'rb') as f:
    wv = pickle.load(f)
all_VIDs = wv.index_to_key
all_VIDs
```

```python
cluster_df_list[-1].index
```

```python
valid_VID_idx = [all_VIDs.index(vid) for vid in cluster_df_list[-1].index]
len(valid_VID_idx)
```

```python
with open(os.path.join(get_data_dir(), "2010s", 'cluster_labels_kmeans.pkl'), 'rb') as f:
    kmeans_labels = pickle.load(f)

cluster_df_list[-1]['kmeans_label'] = kmeans_labels[valid_VID_idx]
```

```python
import numpy as np

with open(os.path.join(get_data_dir(), "2010s", 'cluster_centers_kmeans.pkl'), 'rb') as f:
    kmeans_centers = pickle.load(f)

vectors = wv.get_normed_vectors()[valid_VID_idx]
distances = 1 - np.dot(vectors, kmeans_centers.T)
closest_center_distances = np.min(distances, axis=1)
cluster_df_list[-1]['closest_center_distance'] = closest_center_distances
```

```python
cluster_df_list[-1]
```

```python
cluster_df_list[-1].to_parquet(os.path.join(get_data_dir(), '2010s', 'cluster_df.parquet'))
```
